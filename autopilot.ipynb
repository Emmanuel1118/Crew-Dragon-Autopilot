```python
driver.get("https://iss-sim.spacex.com/") 
wait = WebDriverWait(driver, 600)
```

```python
time.sleep(5) # so i can record it

x_disp = [0,0]
y_disp = [0,0]
z_disp = [0,0]

# displacement values
x_xpath = '//*[@id="x-range"]/div'
x_disp[0] = driver.find_element(By.XPATH, x_xpath).text # getting the string
x_disp[0] = float(str(x_disp[0])[0:len(str(x_disp[0]))-2]) # removing the m
print(x_disp[0])

y_xpath = '//*[@id="y-range"]/div'
y_disp[0] = driver.find_element(By.XPATH, y_xpath).text
y_disp[0] = float(str(y_disp[0])[0:len(str(y_disp[0]))-2])
print(y_disp[0])

z_xpath = '//*[@id="z-range"]/div'
z_disp[0] = driver.find_element(By.XPATH, z_xpath).text
z_disp[0] = float(str(z_disp[0])[0:len(str(z_disp[0]))-2])
print(z_disp[0])

x_disp[1], y_disp[1], z_disp[1] = 0,0,0 # first derivative (rate) initializtion

# rotation values
pitch_xpath_0 = '//*[@id="pitch"]/div[1]'
pitch_xpath_1 = '//*[@id="pitch"]/div[2]'
pitch = [driver.find_element(By.XPATH, pitch_xpath_0).text, driver.find_element(By.XPATH, pitch_xpath_1).text]
pitch[0] = float(pitch[0][0:len(pitch[0])-1]) # displacement
pitch[1] = float(pitch[1][0:len(pitch[1])-4]) # rate
print(pitch)

roll_xpath_0 = '//*[@id="roll"]/div[1]'
roll_xpath_1 = '//*[@id="roll"]/div[2]'
roll = [driver.find_element(By.XPATH, roll_xpath_0).text, driver.find_element(By.XPATH, roll_xpath_1).text]
roll[0] = float(roll[0][0:len(roll[0])-1]) # displacement
roll[1] = float(roll[1][0:len(roll[1])-4]) # rate
print(roll)

yaw_xpath_0 = '//*[@id="yaw"]/div[1]'
yaw_xpath_1 = '//*[@id="yaw"]/div[2]'
yaw = [driver.find_element(By.XPATH, yaw_xpath_0).text, driver.find_element(By.XPATH, yaw_xpath_1).text]
yaw[0] = float(yaw[0][0:len(yaw[0])-1]) # displacement
yaw[1] = float(yaw[1][0:len(yaw[1])-4]) # rate
print(yaw)

#range, the variable is called rang because range is taken by python
range_xpath_0 = '//*[@id="range"]/div[2]'
range_xpath_1 = '//*[@id="rate"]/div[2]'
rang = [driver.find_element(By.XPATH, range_xpath_0).text, driver.find_element(By.XPATH, range_xpath_1).text]
rang[0] = float(rang[0][0:len(rang[0])-2]) # displacement
rang[1] = float(rang[1][0:len(rang[1])-4]) # rate
print(rang)

# inputs

# rotation bottons
pitch_up = driver.find_element(By.XPATH, '//*[@id="pitch-up-button"]')

pitch_down = driver.find_element(By.XPATH, '//*[@id="pitch-down-button"]')

yaw_right = driver.find_element(By.XPATH, '//*[@id="yaw-right-button"]')

yaw_left = driver.find_element(By.XPATH, '//*[@id="yaw-left-button"]')

roll_right = driver.find_element(By.XPATH, '//*[@id="roll-right-button"]')

roll_left = driver.find_element(By.XPATH, '//*[@id="roll-left-button"]')

#translation bottons
trans_up = driver.find_element(By.XPATH, '//*[@id="translate-up-button"]')

trans_down = driver.find_element(By.XPATH, '//*[@id="translate-down-button"]')

trans_right = driver.find_element(By.XPATH, '//*[@id="translate-right-button"]')

trans_left = driver.find_element(By.XPATH, '//*[@id="translate-left-button"]')

trans_frwd = driver.find_element(By.XPATH, '//*[@id="translate-forward-button"]')

trans_bkwd = driver.find_element(By.XPATH, '//*[@id="translate-backward-button"]')

#docking algorithm loop starts here
while(True):
    # storing the previous values in the change values
    x_disp[1], y_disp[1], z_disp[1] = x_disp[0], y_disp[0], z_disp[0]
    #getting all the values, many of the following lines were already written 
    x_disp[0] = driver.find_element(By.XPATH, x_xpath).text # getting the string
    x_disp[0] = float(str(x_disp[0])[0:len(str(x_disp[0]))-2]) # removing the m

    y_disp[0] = driver.find_element(By.XPATH, y_xpath).text
    y_disp[0] = float(str(y_disp[0])[0:len(str(y_disp[0]))-2])

    z_disp[0] = driver.find_element(By.XPATH, z_xpath).text
    z_disp[0] = float(str(z_disp[0])[0:len(str(z_disp[0]))-2])

    # calculating the change
    x_disp[1] = x_disp[0] - x_disp[1]
    y_disp[1] = y_disp[0] - y_disp[1]
    z_disp[1] = z_disp[0] - z_disp[1]

    clear_output(wait=True)
    print(x_disp[1])
    print(y_disp[1])
    print(z_disp[1])

    pitch = [driver.find_element(By.XPATH, pitch_xpath_0).text, driver.find_element(By.XPATH, pitch_xpath_1).text]
    pitch[0] = float(pitch[0][0:len(pitch[0])-1]) # displacement
    pitch[1] = float(pitch[1][0:len(pitch[1])-4]) # rate

    roll = [driver.find_element(By.XPATH, roll_xpath_0).text, driver.find_element(By.XPATH, roll_xpath_1).text]
    roll[0] = float(roll[0][0:len(roll[0])-1]) # displacement
    roll[1] = float(roll[1][0:len(roll[1])-4]) # rate

    yaw = [driver.find_element(By.XPATH, yaw_xpath_0).text, driver.find_element(By.XPATH, yaw_xpath_1).text]
    yaw[0] = float(yaw[0][0:len(yaw[0])-1]) # displacement
    yaw[1] = float(yaw[1][0:len(yaw[1])-4]) # rate

    rang = [driver.find_element(By.XPATH, range_xpath_0).text, driver.find_element(By.XPATH, range_xpath_1).text]
    rang[0] = float(rang[0][0:len(rang[0])-2]) # displacement
    rang[1] = float(rang[1][0:len(rang[1])-4]) # rate

    if roll[0]>0.1: # auto roll
        if roll[0]>10: # for faster rolling in case of big displacement
            if roll[1]<=0.4:
                roll_right.click()
        
        if roll[1]<=0.0:
            roll_right.click()
    
    elif roll[0]<-0.1:
        if roll[0]<-10:
            if roll[1]>=-0.4:
                roll_left.click()
        
        if roll[1]>=0.0:
            roll_left.click()
    
    else:
        if roll[1]<0.0:
            roll_right.click()
        if roll[1]>0.0:
            roll_left.click()

    if pitch[0]>0.1: # auto pitch
        if pitch[0]>10:
            if pitch[1]<=0.4:
                pitch_down.click()
        
        if pitch[1]<=0.0:
            pitch_down.click()
    
    elif pitch[0]<-0.1:
        if pitch[0]<-10:
            if pitch[1]>=-0.4:
                pitch_up.click()

        if pitch[1]>=0.0:
            pitch_up.click()

    else:
        if pitch[1]<0.0:
            pitch_down.click()
        if pitch[1]>0.0:
            pitch_up.click()

    if yaw[0]>0.1: # auto yaw
        if yaw[0]>10:
            if yaw[1]<=0.4:
                yaw_right.click()
        
        if yaw[1]<=0.0:
            yaw_right.click()
    
    elif yaw[0]<-0.1:
        if yaw[0]<-10:
            if yaw[1]>=-0.4:
                yaw_left.click()
        
        if yaw[1]>=0.0:
            yaw_left.click()

    else:
        if yaw[1]<0.0:
            yaw_right.click()
        if yaw[1]>0.0:
            yaw_left.click()

    #approach part
    roll_clear, pitch_clear, yaw_clear = False, False, False

    if abs(roll[0]) < 0.9:
        roll_clear = True

    if abs(pitch[0]) < 0.9:
        pitch_clear = True

    if abs(yaw[0]) < 0.9:
        yaw_clear = True
    
    if roll_clear and pitch_clear and yaw_clear:
        print('clear')
        if z_disp[0] > 0.1: # z
            if z_disp[1] >= 0.0:
                trans_down.click()

        elif z_disp[0] < -0.1:
            if z_disp[1] <= 0.0:
                trans_up.click()

        else:
            if z_disp[1] > 0.0:
                trans_down.click()
            elif z_disp[1] < 0.0:
                trans_up.click()

            '''else:
                if z_disp[0]>0.0:
                    trans_down.click()
                elif z_disp[0]<0.0:
                    trans_up.click()'''

        if y_disp[0] > 0.1: # y
            if y_disp[1] >= 0.0:
                trans_left.click()

        elif y_disp[0] < -0.1:
            if y_disp[1] <= 0.0:
                trans_right.click()

        else:
            if y_disp[1] > 0.0:
                trans_left.click()
            elif y_disp[1] < 0.0:
                trans_right.click()

            '''else:
                if y_disp[0]>0.0:
                    trans_left.click()
                elif y_disp[0]<0.0:
                    trans_right.click()'''

        if x_disp[0]>50: # initial range
            if rang[1]>-1.1: # min speed
                trans_frwd.click()

            if rang[1]<-1.3: # max speed
                trans_bkwd.click()

        elif x_disp[0]<50 and x_disp[0]>10: # second phase
            if rang[1]>-0.4: # min speed
                trans_frwd.click()

            if rang[1]<-0.6: # max speed
                trans_bkwd.click()

        else:
            if abs(y_disp[0]) < 2 and abs (z_disp[0]) < 2: # final phase, if alligned
                if rang[1]>-0.04: # min speed
                    trans_frwd.click()

                if rang[1]<-0.12: # max speed
                    trans_bkwd.click()

            else: # if not alligned, stop at 10m
                if x_disp[0]>10.0:
                    trans_frwd.click()

                if x_disp[0]<10.0:
                    trans_bkwd.click()

# when the game ends the program just crashes lol
```
